### Gradient descent draft

```python
X = train.iloc[:, lambda df: [0,1,2,3,4,5,6,7,8]]
Y = train.iloc[:, lambda df: [9]]
```

```python
m = 0
c = 0

l_rate = 0.0001  # The learning Rate
num_iterations = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(num_iterations): 
    Y_pred = m*X + c  # The current predicted value of Y
    dm = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    dc = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - l_rate * dm  # Update m
    c = c - l_rate * dc  # Update c
    
print (m, c)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
   # y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / (y_true+1e-10)))
```
